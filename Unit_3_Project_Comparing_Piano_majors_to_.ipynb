Now to calculate the true percentage...

convert percentage to **fractions**...

Piano = 5.0% = 1/20

Percussion = 3.4% = 17/500

Violin = 4.5% = 9/200

Guitar = 4.5% = 9/200

**add** the fractions up...

1/20 + 17/500 + 3(9/200) = 0.219 = 21.9%


**calculate** percentage of piano from percentage of all...

0.05/0.219 = 0.228 = 22.8%
